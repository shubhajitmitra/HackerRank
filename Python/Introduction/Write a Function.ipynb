# Leap year
   - **Problem Difficulty -  Medium**

```python
def is_leap(year):
    leap = False

    # Write your logic here
    if year<=1900 and year>=10**5:
        return leap
    elif year %100 == 0 and year %400 != 0:
        return leap
    elif year%4 !=0:
        return leap
    else:
        return True

year = int(input())
print(is_leap(year))
```
