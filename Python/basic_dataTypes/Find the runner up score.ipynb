# Find a runner up score
   - **Problem Class - Easy** 

```python
if __name__ == '__main__':
    n = int(input())
    arr = map(int, input().split())
    score_list = list(arr)[:n]
    highest_score = max(score_list)
    while max(score_list)==highest_score:
        score_list.remove(highest_score)
    print(max(score_list))
```
