# List
- **Problem Difficulty - Easy**

```python
if __name__ == '__main__':
    N = int(input())   
    empty_list = []
    converted_list = []

    for i in range(N):
        x = input().split()
        empty_list.append(x)

    for i in range(len(empty_list)):
        if empty_list[i][0] == 'insert':
            x = int(empty_list[i][1])
            y = int(empty_list[i][2])
            converted_list.insert(x,y)
        elif empty_list[i][0] == 'print':
            print(converted_list)
        elif empty_list[i][0] == 'remove':
            converted_list.remove(int(empty_list[i][1]))
        elif empty_list[i][0] == 'append':
            converted_list.append(int(empty_list[i][1]))
        elif empty_list[i][0] == 'sort':
            converted_list.sort()
        elif empty_list[i][0] == 'pop':
            converted_list.pop()
        elif empty_list[i][0] == 'reverse':
            converted_list.reverse()
```
