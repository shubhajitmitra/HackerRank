```python
# People who got Second Lowest

if __name__ == '__main__':
    mc = []
    grade = []
    for _ in range(int(input())):
        student_name = input()
        score = float(input())
        mc.append([student_name,score])
        sorted_mc = sorted(mc)
        grade.append(score) 
        
    sorted_uniq_grade = sorted(list(set(grade)))
    second_lowest_grade = (sorted_uniq_grade)[1]
        
    for i in (sorted_mc):
        if i[1] == second_lowest_grade:
            print(i[0])
            continue     
```

```python
# People who got Second Highest

if __name__ == '__main__':
    mc = []
    grade = []
    for _ in range(int(input())):
        student_name = input()
        score = float(input())
        mc.append([student_name,score])
        sorted_mc = sorted(mc)
        grade.append(score)
    sorted_uniq_grade = sorted(list(set(grade)))
    second_highest_grade = sorted_uniq_grade[-2:][0] # [-2:] will select last two elements of the list and then[0] will select first element of those two elements 
        
    for i in (sorted_mc):
        if i[1] == second_highest_grade:
            print(f"Second highest grades are of {i[0]}")
            continue        
```
