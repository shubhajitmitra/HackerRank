```python
# mutatate a string by changing a charecter (with out changing its length)

def mutate_string(string, position, character):
    mutated_string = string[:position]+str(character)+string[position+1:]
    return mutated_string

if __name__ == '__main__':
    s = input()
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
```

```python
# mutatate a string by inserting a charecter at a perticular position. This will changing its length.
def mutate_string(string, position, character):
    mutated_string = string[:position]+str(character)+string[position:]
    return mutated_string

if __name__ == '__main__':
    s = input()
    i, c = input().split()
    s_new = mutate_string(s, int(i), c)
    print(s_new)
```
